## **5️⃣ Mapping Data Type (dict)**
A Mapping Data Type stores data in the form of key-value pairs.

Each key is unique and is associated with a specific value.



A **Dictionary (dict)** is a built-in Python data type that stores data as key-value pairs, allowing fast retrieval of values using their keys.



**Why Do We Need Dictionaries?**

Suppose you want to store a student's information.

### Using separate variables:

    name = "Anjali"
    age = 22
    course = "B.Sc CS"
    city = "Haldwani"

Managing many variables becomes difficult.

**Using a dictionary:**

    student = {
      "name": "Anjali",
      "age": 22,
      "course": "B.Sc CS",
      "city": "Haldwani"
     }

Now all related information is stored in one place.



**Syntax**

A dictionary is created using curly braces {}.

    dictionary_name = {
    "key1": value1,
    "key2": value2,
    "key3": value3
     }



     

```python
#Example
student={
    "name":"Anjali",
    "age":22,
    "course":"MBA"
}

print(student)
```

```python
print(type(student))
```

```python
employee= {
    "id":101,
    "name":"Anjali",
    "department":"Analytics",
    "Salary":95000
}
print(employee)
```

**Characteristics of Dictionary**

*  Stores data as key-value pairs.

*  Keys must be unique.

*  Values can be duplicated.

*  Mutable (values can be changed).

*  Fast access using keys.

*  Can store different data types.

## **6️⃣ Set Data Type (set)**


A **Set** (set) is a built-in Python data type used to store a collection of unique elements enclosed in curly braces **{}**.

 Unlike lists and tuples, a set automatically removes duplicate values.

**Syntax**

A set is created using **curly braces {}**.

    set_name = {item1, item2, item3}

### **Why Do We Need Sets?**

Imagine we have a list of student IDs with duplicates.

    student_ids = [101, 102, 103, 101, 102]

If you only want unique IDs, a set is the perfect choice.

    student_ids = {101, 102, 103}

Sets help remove duplicate values automatically.





```python
fruits ={"Apple", "Mango", "Banana"}
print(fruits)
#Note: The order of elements may vary because sets are unordered.
```

```python
print(type(fruits))
```

## **Characteristics of Set**
Stores only unique elements.

Automatically removes duplicate values.

Unordered collection (do not rely on element order).

Mutable (you can add or remove elements).

Does not support indexing or slicing.

Can store different data types.

### **Duplicate Values Example**

```python
number = {10,20,30,40,10,40}
```

```python
print(number)
```

## **7️⃣ None Type (NoneType)**

None is a special constant in Python that represents the absence of a value or no value.

The data type of **None** is called **NoneType**.

**Syntax**

    variable_name = None

**Why Do We Need None?**

Sometimes, a variable is created but no value is assigned yet.

**For example:**

*  A student's marks are not available.
*  A customer has not entered their phone number.
*  A function does not return any value.

In these situations, we use **None**.


```python
#Example
result=None
print(result)
```

```python
print(type(result))
```

## **Characteristics of None**

Represents the absence of a value.

There is only one None object in Python.

Data type is NoneType.

None is not the same as 0, False, or an empty string ("").

Commonly used as a default value in functions and variables.


**Example**

Imagine an online shopping application.

    None → No discount has been assigned.

    0 → Quantity is zero.

    False → Payment has not been completed.

    "" → Coupon code is empty.

    [] → Cart exists but contains no products.

# **What is Indexing?**

Indexing is the process of accessing an individual element or character from a sequence using its index number.


**Why Do We Need Indexing?**

Suppose you have the following string:

    language = "Python"

If you want only the first letter "P", you don't need the entire string.

**Using indexing:**

    print(language[0])

**Output:**

     P

Without indexing, it would be impossible to directly access individual elements.


**Characteristics of Indexing**

Starts from 0.

Every element has a unique index.

Supports both positive and negative indexing.

Works only with ordered data types.

Allows direct access to any element.




Data Types That Support Indexing


String (str) --	✅ Yes

List (list) --	✅ Yes

Tuple (tuple)	  -- ✅ Yes

Range (range) --	✅ Yes

Dictionary (dict) --	❌ No

Set (set)	--❌ No





# **Types of Indexing**

Python supports two types of indexing:



*   Positive Indexing
*    Negative Indexing


## **1️⃣ Positive Indexing**

Positive indexing starts counting from the left.

The first element always has **index 0**.

    String :   P    Y    T    H    O    N
    Index  :   0    1    2    3    4    5


    


```python
#Example
language ="Python"
print(language[0])
```

```python
print(language[1])
```

```python
# positive indexing
fruits =["Apple", "Mango", "Banana"]
print(fruits[0])
```

```python
color = ("red", "green", "blue")
print(color[2])
```

# **2️⃣ Negative Indexing**

Negative indexing starts counting from the right.

The last element always has **index -1**.

    String :    P    Y    T    H    O    N
    Index  :   -6   -5   -4   -3   -2   -1


```python
language = "Python"

print(language[-1])
print(language[-2])
print(language[-3])
print(language[-6])
```

```python
numbers=[10,30,20,89,58]
print(numbers[-1])
```

```python
print(numbers[-3])
```

# **Mutability and Immutability**
**What is Mutability?**

Mutability is the ability of an object to change or modify its contents after it has been created.

If an object can be changed without creating a new object, it is called **mutable**.


**What is Immutability?**

Immutability is the property of an object whose value cannot be changed after it has been created.



**Simple Meaning**

**Mutable**  --- Can be changed after creation

**Immutable** ---	   Cannot be changed after creation



# **Real-Life Analogy**
📝 **Mutable Object**→ Whiteboard

A whiteboard can be written on, erased, and updated many times.


**Immutable Object** → Printed Book

Once a book is printed, you cannot change its contents.

If changes are needed, a new edition must be printed.

```python
# Example of Mutable objects
fruits = ["Apple", "Mango", "Banana"]
print(fruits)
```

```python
fruits[1]
```

```python
fruits[0]="Orange"
```

```python
print(fruits) #The list was modified without creating a new object.
```

```python
#Example of an Immutable Object
name = "anjali"
```

```python
name = "Rohan"
print(name)
#Explanation
#The original string "Anjali" was not modified.
#Python created a new string object containing "Rohan"
```
