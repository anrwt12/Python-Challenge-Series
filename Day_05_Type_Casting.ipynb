# **What is Type Casting?**


Type Casting is the process of converting a value from one data type to another data type.

It is also known as **Type Conversion**.

| Original Data Type | Converted Data Type |
| ------------------ | ------------------- |
| String             | Integer             |
| Integer            | Float               |
| Float              | Integer             |
| Integer            | String              |
| List               | Tuple               |
| Tuple              | Set                 |

---

**Real-Life Analogy**

Imagine you are travelling from India to Japan.

In **India**, you use **Indian Rupees (₹)**,

 but in **Japan**, shops accept only **Japanese Yen (¥)**.

Before you can buy anything, you need to **convert your money** into the local currency.


**Indian Rupees (₹)**-->
**Currency Exchange**-->
**Japanese Yen (¥)**

Similarly, in Python, data sometimes needs to be converted into another data type before it can be processed correctly.

**String ("25")-->
Type Casting-->
Integer (25)**

Just as currency conversion allows you to use money in another country,


Type Casting allows Python to use data in the required format.


```python
# here the number is  stores in string

number = "100"
print(number)
print(type(number))
```

```python
number = int(number) # convert the string into a integer
print(number)# after conversion python treat it as a numeric value
print(type(number))
```

```python
salary = 80000
print(type(salary))

```

```python
salary = float(salary)
print(salary)
print(type(salary))
```

```python
#Working with User Input
price = float(input("Enter the product price:"))

quality = int(input("Enter the product quality:"))

bill = price * quality

print("total_bill =", bill)
```

# **Types of Type Casting**

Python supports two types of Type Casting.

**Implicit Type Casting** (Automatic Type Conversion)

**Explicit Type Casting** (Manual Type Conversion)

---

# **Implicit Type Casting (Automatic Type Conversion)**

Implicit Type Casting is the automatic conversion of one data type into another by Python.

```python
num1 = 10 # --> integer
num2 = 2.2 #--> float
result = num1+ num2
print(result) # final answer becomes a float
print(type(result))
```

```python
a = 20
b= 2.5
c= a*b
print(c)
```

**2️⃣ Explicit Type Casting (Manual Type Conversion)**


Explicit Type Casting is the process of manually converting one data type into another using Python's built-in conversion functions.

```python
age = "21"
print(type(age))
```

```python
age = int(age)
print(age)
print(type(age))
```

```python
#int() Function
#Example 1: String to Integer
number = "100"
print("Before Conversion:", number)
print("Data type :",type(number))

```

```python
number = int(number)
print("After Conversion:", number)
print("Data type :",type(number))
```

```python
#Example 2: Float to Integer
price = 99.99
price = int(price)
print(price)
print(type(price))



```

```python
#2. float() Function
#Example 1: Integer to Float
num = 25
num = float(num)
print(num)
print(type(num))
```

```python
#Example 2: String to Float
price = "100.5"
price = float(price)
print(price)
print(type(price))
```

```python
#3. str() Function
#Example 1: Integer to String
age = 21
age = str(age)
print(age)
print(type(age))
```

```python
#Example 2: Float to String
price = 999.99
price = str(price)
print(price)
print(type(price))
```

```python
#4. bool() Function
#Example 1: Integer to Boolean

print(bool(0))
print(bool(1))
```

```python
#Example 2: String to Boolean

print(bool("python"))
print(bool("python"))
```

```python
#5. list() Function
#Example 1: String to List
course = "python"
course = list(course)
print(list("python"))
print(type(course))

```

```python
#Example 2: Tuple to List
numbers=(10,20,30,40)
numbers= list(numbers)
print(numbers)
print(type(numbers))
```

```python
#6. tuple() Function
#Example 1: List to Tuple
fruits = ["apple", "banana", "cherry"]
fruits = tuple(fruits)
print(fruits)
print(type(fruits))
```

```python
#Example 2: String to Tuple
word = "Python"
word = tuple(word)
print(word)
print(type(word))
```

```python
#7. set() Function
#Example 1: List to Set
numbers = [10, 20, 30, 20, 10]
numbers = set(numbers)
print(numbers)
print(type(numbers))
```

```python
#Example 2: String to Set
word = "Python"
word = set(word)
print(word) #A set is unordered, so the order of elements may vary.
print(type(word))
```

```python
#8. dict() Function
#Example 1: List of Tuples to Dictionary
student=[("name","Anjali"),("Age",21)]
student=dict(student)
print(student)
print(type(student))
```

```python
#9. Type Casting with User Input (input())
#Example 1: Addition of Two Numbers
num1 = int(input("Enter First Number: "))
num2 = int(input("Enter Second Number: "))
sum = num1 + num2
print("Sum =", sum)
```

```python
#Example 2: Calculate Area of a Circle
radius = float(input("Enter the radius of the circle: "))
area = 3.14 * radius * radius
print("Area of the circle is:", area)
```
