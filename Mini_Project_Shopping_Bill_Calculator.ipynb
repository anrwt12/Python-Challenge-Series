# **📝 Problem Statement**

Create a Shopping Bill Calculator that:

* Accepts shop and customer details.
* Accepts details of two products.
* Calculates the total price for each product.
* Calculates the grand total.
* Displays the data types of important variables.
* Prints a professional invoice.

```python
# ==========================================================
#              MINI PROJECT : SHOPPING BILL CALCULATOR
# ==========================================================

# -----------------------------
# Shop Information
# -----------------------------
shop_name = "Läderach"
shop_owner = "Anjali"
shop_location = "Bern,Switzerland"

# -----------------------------
# Welcome Message
# -----------------------------

print("=" * 60)
print("\t\tWELCOME TO",shop_name.upper())
print("=" * 60)
print("Address : =",shop_location)
print()

# -----------------------------
# Customer Information
# -----------------------------
customer_name = input("Enter Your Name :")
mobile_number = input("Enter Your Mobile Number :")
print()

# -----------------------------
# Product Details
# -----------------------------

product_name_1 =  str(input("Enter product 1 name :"))
product_price_1= input("Enter product price :")
quantity_1= input("Enter Qunatity :")
print()
product_name_2 =  str(input("Enter product 2 name :"))
product_price_2=input("Enter product price :")
quantity_2 = input("Enter Qunatity :")
print()

# -----------------------------
# Type Casting
# -----------------------------
product_price_1 = float(product_price_1)
product_price_2 = float(product_price_2)
quantity_1= int(quantity_1)
quantity_2= int(quantity_2)

# -----------------------------
# Calculations
# -----------------------------
subtotal_1 = product_price_1 * quantity_1
subtotal_2 = product_price_2 * quantity_2
grand_Total = subtotal_1 + subtotal_2

# -----------------------------
# Ending Message
# -----------------------------
print("=" *60)
print("\t\tWELCOME TO",shop_name.upper())
print("=" * 60)
print("Customer Name :", customer_name)
print("Mobile Number :", mobile_number)
print("Product 1 Name :", product_name_1)
print("Product 1 Subtotal :", subtotal_1)
print("Product 2 Name :", product_name_2)
print("Product 2 Subtotal :", subtotal_2)
print("Grand Total :", grand_Total)

print("Thank You for Shopping!")
print("Visit Again 😊")
print("=" * 60)



```
