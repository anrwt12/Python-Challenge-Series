# **What are Comments?**

Comments are notes written inside a Python program to explain the code. They are meant for humans to read, not for the computer to execute.

**Why Do We Use Comments?**

Comments are useful because they help us

✅ Explain what the code does.

✅ Improve code readability.

✅ Make debugging easier.

✅ Help other programmers understand the code.

✅ Document complex logic.

✅ Remember the purpose of a section of code.



### **Types of Comments in Python**

Python supports two main types of comments:

1️⃣ **Single-Line Comment**

A single-line comment starts with the hash (**#**) symbol.

Everything after **#** on that line is treated as a comment.

```python
# syntax
#This is a comment
```

```python
#print a welcome message
print("welcome to python")
```

```python
# Program to calculate the sum of two numbers
# Store values in variables
a = 10
b = 20

# Display the result
print(a + b)
```

# **2️⃣ Multi-Line Comment**

Python does not have a dedicated multi-line comment syntax. However, programmers often use triple quotes (**''' '''** or **""" """**) to write multi-line descriptions.

```python
"""
This  program demonstrates
how to use
multi-line comments.
"""
print("hello world")
```

# **6️⃣ Print End (end)**

By default, **print()** moves the cursor to the next line after printing.

The **end** parameter changes this behavior.

```python
# Example
print("hello" , end="__")
print("World")
```

```python
print("hi i am", end="-->")
print("Anjali")
```
