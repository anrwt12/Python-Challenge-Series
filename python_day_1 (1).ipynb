# ** Q1 What is programming language?**

A programming language is a language used to communicate with a computer. It allows us to write **instructions** that tell the computer what tasks to perform.

## ** Real-Life Example**

Imagine you are making a cup of tea. You follow these steps:

Boil the water.

Add tea leaves.

Add sugar.

Add milk.

Pour the tea into a cup.

These steps are called **instructions**.

Similarly, a computer also needs step-by-step instructions to perform a task. These instructions are written using a programming language.

**Example**

    print("Hello, World!")

# **Python Syntax**

Every programming language has a set of rules called syntax.

Syntax tells us how to write code correctly.

One of the biggest advantages of Python is that its syntax is clean and easy to understand.


## Example

    print("Hello, World!")



1️⃣ **Python is Case-Sensitive**

    name = "Anjali"
    Name = "Python"

Here, **name** and **Name **are considered different variables.


2️⃣ **Indentation Matters**

Python uses indentation (spaces) to define blocks of code.

    if 10 > 5:
        print("10 is greater than 5")

If the indentation is incorrect, Python will generate an error.

3️⃣ **No Semicolon Needed**

Unlike some other programming languages, Python does not require a semicolon at the end of each statement.

    print("Python")
    print("Programming")

4️⃣ **Strings Can Use Single or Double Quotes**

      print("Hello")
      print('Python')

Both are valid.
