# **What are Operators?**

An Operator is a special symbol or keyword that tells Python to perform a specific operation on one or more values.


```python
a= 10
b= 20
print(a+b)

"""
a is an operand having value 10
b is an operand having value 20
+ is an operator
+ tells Python to add the two operands
"""
```

# **Why are Operators Important?**

Without operators, Python cannot perform calculations or make decisions.

Operators are used in almost every Python program.

**They help us to**:

Perform mathematical calculations

Compare two values

Assign values to variables

Check logical conditions

# **4. Logical Operators**

Logical operators are used to combine two or more conditions into a single expression.

They are widely used in conditional statements and decision-making.

**Python provides three logical operators:**
*  and
*   or
*   not




| Operator | Name        | Purpose                                              |
| -------- | ----------- | ---------------------------------------------------- |
| `and`    | Logical AND | Returns `True` if **all conditions** are True        |
| `or`     | Logical OR  | Returns `True` if **at least one condition** is True |
| `not`    | Logical NOT | Reverses the Boolean result                          |


 **Logical AND Operator (and)**


The **AND** operator returns True only when all the given conditions are True.

If even one condition is False, the result becomes **False**.

**Syntax**

    ondition1 and condition2

 | Condition 1 | Condition 2 | Result |
| ----------- | ----------- | ------ |
| True        | True        | True   |
| True        | False       | False  |
| False       | True        | False  |
| False       | False       | False  |


```python
age = 22
salary = 80000
print(age >= 18 and salary >=50000)
```

```python
marks = 75
attendance = 70
print(marks >= 70 and attendance >= 75)

```

**Logical OR Operator (or)**


The **OR** operator returns **True if at least one** **condition is**  **True**.

It returns False only when all conditions are False.

**Syntax**

    condition1 or condition2

| Condition 1 | Condition 2 | Result |
| ----------- | ----------- | ------ |
| True        | True        | True   |
| True        | False       | True   |
| False       | True        | True   |
| False       | False       | False  |



```python
age = 16
parent_permission = True

print(age >= 18 or parent_permission)
```

```python
member = False
purchase = 1500
print(member or purchase >= 1000)
```

**Logical NOT Operator (not)**


The **NOT** operator is used to reverse the result of a Boolean expression.

*  True becomes False

*  False becomes True

**Syntax**


    not condition

| Condition | Result |
| --------- | ------ |
| True      | False  |
| False     | True   |


```python
is_logged_in = False
print(not is_logged_in)
```

```python
age = 16
print(not(age>=18))
```

# **5. Identity Operators**

Identity operators are used to determine whether two variables refer to the same object in memory.

These operators compare the identity of objects rather than their values.

**Operators**

*  is
*  is not


| Operator | Name                  | Purpose                                                     |
| -------- | --------------------- | ----------------------------------------------------------- |
| `is`     | Identity Operator     | Returns `True` if both variables refer to the same object   |
| `is not` | Identity Not Operator | Returns `True` if both variables refer to different objects |


**Identity vs Equality**

| Identity (`is`)                                           | Equality (`==`)                    |
| --------------------------------------------------------- | ---------------------------------- |
| Compares memory location                                  | Compares values                    |
| Checks object identity                                    | Checks value equality              |
| Returns `True` if both variables point to the same object | Returns `True` if values are equal |


```python
a =[10,20,30]
b=[10,20,30]
print(a is b)
#a is b → Both lists are stored at different memory locations, so the result is False.

```

```python
print(a == b)
#a == b → Both lists have the same values, so the result is True.
```

**Identity Operator (is)**


The **is** operator checks whether two variables refer to the **same object in memory**.

If both variables point to the same object, it returns **True**; otherwise, it returns **False**.

**Syntax**

    object1 is object2

```python
a= [ 1,2,3,4]
b = a
print(a is b)
```

```python
print(b is a)
```

```python
x= 100
y=100
print(x is y)
```

**Identity Not Operator (is not)**


The **is not** operator checks whether two variables refer to different objects in memory.

It returns **True** if the objects are different.

**Syntax**

    object1 is not object2

```python
a = [5,10]
b = [5,10]
print(a is not b)
```

# **6. Membership Operators**

Membership operators check whether a particular element exists in a sequence such as a list, tuple, string, set, or dictionary.

**Operators**
*  in
*  not in


| Operator | Name                         | Purpose                                                        |
| -------- | ---------------------------- | -------------------------------------------------------------- |
| `in`     | Membership Operator          | Returns `True` if the element exists in the collection         |
| `not in` | Negative Membership Operator | Returns `True` if the element does not exist in the collection |


**Membership Operator (in)**


The **in** operator checks whether a specified element is present in a collection.

If the element is found, it returns **True**; otherwise, it returns **False**.

**Syntax**

    element in collection

```python
fruits = ["Apple","Banana","Orange"]
print("Apple" in fruits)
```

```python
language = "Python"
print("P" in language)
```

```python
numbers =(10, 20 , 30 ,40)
print(20 in  numbers)
```

```python
colors = {"Red","Green","Blue"}
print("Yellow" in colors)
```

**Membership Operator (not in)**


The **not in** operator checks whether an element is not present in a collection.

It returns **True** when the element does not exist.

**Syntax**

    element not in collection

```python
fruits = ["Apple", "Banana", "Orange"]

print("Mango" not in fruits)
```

```python
word = "Python"

print("Z" not in word)
```

```python
numbers = (1, 2, 3, 4)

print(10 not in numbers)
```

# **7. Bitwise Operators**

Bitwise operators work on the binary representation of integers.

These operators are mainly used in low-level programming, networking, encryption, and hardware-related applications.


**Types of Bitwise Operators**



| Operator | Name                       | Meaning                                                                                                                                      |
| -------- | -------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------- |
| `&`      | Bitwise AND                | Returns **1** if **both bits are 1**, otherwise returns **0**.                                                                               |
| `\|`     | Bitwise OR                 | Returns **1** if **at least one bit is 1**, otherwise returns **0**.                                                                         |
| `^`      | Bitwise XOR (Exclusive OR) | Returns **1** if the corresponding bits are **different**, otherwise returns **0**.                                                          |
| `~`      | Bitwise NOT                | Inverts (flips) all the bits (**1 → 0**, **0 → 1**).                                                                                         |
| `<<`     | Left Shift                 | Shifts all bits to the **left** by the specified number of positions (approximately multiplies by **2** for each shift).                     |
| `>>`     | Right Shift                | Shifts all bits to the **right** by the specified number of positions (approximately divides by **2** for each shift for positive integers). |



```python
a = 5
b = 3

print(a & b)   # 1
print(a | b)   # 7
print(a ^ b)   # 6
print(~a)      # -6
print(a << 1)  # 10
print(a >> 1)  # 2
```
